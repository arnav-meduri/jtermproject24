```python
arrest_counts = data['ARREST'].value_counts()

domestic_counts = data['DOMESTIC'].value_counts()

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

axes[0].bar(arrest_counts.index, arrest_counts.values, color=['blue', 'blue'])
axes[0].set_title('Number of Arrested vs Non-Arrested')
axes[0].set_xlabel('Arrested')
axes[0].set_ylabel('Number of Cases')

axes[1].bar(domestic_counts.index, domestic_counts.values, color=['red', 'red'])
axes[1].set_title('Number of Domestic vs Non-Domestic Cases')
axes[1].set_xlabel('Domestic')
axes[1].set_ylabel('Number of Cases')

# Displaying the plots
plt.tight_layout()
plt.show()
```

```python
# Correcting the column name for 'PRIMARY DESCRIPTION' (removing leading space)
crime_counts = data[' PRIMARY DESCRIPTION'].value_counts()

# Sorting the crimes by frequency
sorted_crime_counts = crime_counts.sort_values(ascending=False)

# Creating the bar plot
plt.figure(figsize=(10, 8))
sns.barplot(x=sorted_crime_counts.values, y=sorted_crime_counts.index, palette="vlag")

plt.title('Number of Each Type of Crime', fontsize=16)
plt.xlabel('Number of Cases', fontsize=12)
plt.ylabel('Type of Crime', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Displaying the plot
plt.show()
```
